```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from youtube import YouTube
```

```python
#read API key
def read_api_key(filepath):
    """Read our API key from provided filepath
    
    Args:
        filepath (string): File containing the API key
        
    Returns
        api_key (string): the API key
        
    """
    with open(filepath, 'r') as f:
        return f.read().replace('\n', '')
```

```python
# connect to youtube API
API_key = read_api_key('youtube_api_key.txt')
yt = YouTube(API_key)
```

# Get Videos from YouTube API

```python
#send to query to YouTube API and get search results
num_videos = 500
search_result = yt.get_search_results(query='humpback whale', license_type = 'creativeCommon', total_num_videos=num_videos)
len(search_result)
```

```python
#store search result into dataframe
cc_youtube_videos = pd.DataFrame(search_result)
cc_youtube_videos['relevant'] = [None for i in range(0, len(cc_youtube_videos))]

cc_youtube_videos.to_csv('/Users/micheller/Documents/nvidia-intern-project/data/creative_commons_youtube_videos.csv', index=False)
```

# Load Saved DF 

```python
#load videos in from saved df
cc_youtube_videos = pd.read_csv('/Users/micheller/Documents/nvidia-intern-project/data/creative_commons_youtube_videos.csv')
cc_youtube_videos
```

# Label Videos

```python
#go through search results and label videos, storing info for those you keep

print('How many videos do you want to filter?:', end = " ")
num_videos_to_filter = int(input())

start_filtering_at_df_index = cc_youtube_videos[cc_youtube_videos['relevant'].isna()].index[0]
end_filtering_at_df_index = start_filtering_at_df_index + num_videos_to_filter

for i, row in cc_youtube_videos[start_filtering_at_df_index: end_filtering_at_df_index].iterrows():
    print(f'Index {i}')
    relevant_label = yt.label_video(row['video_id'])
    cc_youtube_videos.at[i, 'relevant'] = relevant_label

print(f'Done filtering videos from index {start_filtering_at_df_index} - {end_filtering_at_df_index}')
```
