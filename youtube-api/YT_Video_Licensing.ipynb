```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
from youtube import YouTube
```

## Get license info for each video

```python
#read API key
def read_api_key(filepath):
    """Read our API key from provided filepath
    
    Args:
        filepath (string): File containing the API key
        
    Returns
        api_key (string): the API key
        
    """
    with open(filepath, 'r') as f:
        return f.read().replace('\n', '')
```

```python
# connect to youtube API
API_key = read_api_key('youtube_api_key.txt')

yt = YouTube(API_key)
```

```python
#load in data
csv_path = '/Users/micheller/Documents/nvidia-intern-project/data/youtube_humpback_whale_classification_dataset.csv'
url_data = pd.read_csv(csv_path)

#get license information about each video
legal_info = dict()
for url in url_data['url']:
    video_id = url.replace('https://youtu.be/', '')
    result = yt.videos(video_id)

    try:
        legal_info[video_id] = {'url': url, 
                                'privacyStatus': result[0]['status']['privacyStatus'],
                                'licensedContent':result[0]['contentDetails']['licensedContent'],
                                'license': result[0]['status']['license']}
    except IndexError:
        legal_info[video_id] = {'url': url, 
                                'privacyStatus': 'unavailable',
                                'licensedContent':'unavailable',
                                'license': 'unavailable'}
    
    time.sleep(0.5)

legal_info
```

```python
legal_df = pd.DataFrame.from_dict(legal_info, orient='index')
legal_df 
```

```python
#save legal data
legal_df.to_csv('/Users/micheller/Documents/nvidia-intern-project/data/youtube_whale_videos_legal_info.csv')
```

## Plot counts of each type of license retrieved for our videos

```python
# licensedContent counts
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize':(10,5)})
plt.figure(dpi=100)
ax = legal_df['licensedContent'].value_counts().plot(kind='barh', title='Licensed Content Status of YouTube Whale Videos');

for i, v in enumerate(legal_df['licensedContent'].value_counts()):
    ax.text(v + 3, i, str(v), color='black', fontsize = 10)

plt.savefig('/Users/micheller/Documents/nvidia-intern-project/figures/video_licensed_content.png')
```

```python
# license counts
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize':(10,5)})
plt.figure(dpi=100)
ax= legal_df['license'].value_counts().plot(kind='barh', title='Types of Licenses of YouTube Whale Videos');

for i, v in enumerate(legal_df['license'].value_counts()):
    ax.text(v + 3, i, str(v), color='black', fontsize = 10)

plt.savefig('/Users/micheller/Documents/nvidia-intern-project/figures/video_licenses.png')
```

```python
# privacy status counts
sns.set_theme(style='darkgrid')
sns.set(rc={'figure.figsize':(10,5)})
plt.figure(dpi=100)
ax= legal_df['privacyStatus'].value_counts().plot(kind='barh', title='Privacy Status of YouTube Whale Videos');

for i, v in enumerate(legal_df['privacyStatus'].value_counts()):
    ax.text(v + 3, i, str(v), color='black', fontsize = 10)

plt.savefig('/Users/micheller/Documents/nvidia-intern-project/figures/video_privacy_status.png')
```

`note: "unavailable" denotes videos that were once available but are no longer publicly accessible (either deleted by user, or user changed to private listing)`
